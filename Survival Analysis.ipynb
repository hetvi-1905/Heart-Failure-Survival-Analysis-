## Modelling Approaches


### 1) KaplanMeier Estimator

### The Kaplan-Meier estimator
##### The Kaplan-Meier estimator is a test statistic that gives us an approximation of the true survival function of a population, the approximation getting better with increasing sample size. This estimator can robustly handle censoring (incomplete data) , and can be derived from the Hazard Function using Maximum Likelihood Estimation.

This estimator can be used for simple comparison of survival rates between "groups" (categorical data ) (For instance, survival rates between smokers and non-smokers).

Its mostly used . Its a non parametric method to estimate survival functions.Byy Non parametric it means that it doesn't assume any particular underlying distribution (like normal, exponential dist.) for survival times, neither does it rely on params like mean and variance to model survival data.(makes no assumption about the shape of survival courve

### 2)  Cox Proportional Hazard Model

The Cox Proportional Hazard Model is a survival analysis model that assumes that the baseline hazard function of a population is multiplicatively influenced by the covariates. For instance, given a group of smokers and non-smokers, the baseline hazard of the population is multiplied by 2 in case of smokers.

The Cox Proportional Hazard model can be used to model hazard rates based on multiple features, either categorical or numerical.

λ(t|x)=λ0(t)eb1x1+b2x2+...+bnxn In the above equation:

λ0(t) is the baseline hazard function of the population. It varies with time, but is otherwise independent of the features. Each feature (the x's) multiplicatively affect the baseline hazard function (multiplying by ebkxk ) The term "proportional hazard" means that all feature effects are proportional throughout time. If a certain feature doubles the baseline hazard rate at time t = 2, then it also doubles the baseline hazard rate at time t = 3, t = 100 etc.

### Difference between both the methods ?? And does it means that kaplan method works on categorical data only and Cox Proportional Hazard Model on both categorical and numerical data ?




Not exactly. The difference between the Kaplan-Meier method and the Cox proportional hazards model is more about how they handle time-to-event data and covariates, rather than being restricted to categorical or numerical data.

Kaplan-Meier Method:
Primarily focuses on time-to-event data (like survival time), without considering additional covariates.
It is often used to compare survival between groups (categorical data), such as treatment vs. control groups. For example, you can compare survival curves for two or more categories (e.g., male vs. female, different treatment types).

However, the Kaplan-Meier method doesn't handle continuous covariates (e.g., age, blood pressure, etc.) well. You would typically categorize continuous variables to use them with this method (e.g., dividing age into age groups).


Cox Proportional Hazards Model:
Handles both categorical and continuous (numerical) covariates. This is one of its strengths.
For example, you can include both categorical variables (e.g., treatment group, gender) and continuous variables (e.g., age, blood pressure, tumor size) to see how they affect the survival time.
It models the relationship between survival and the covariates, estimating how these factors change the hazard (risk) of an event occurring.

Summary:
Kaplan-Meier works primarily for categorical data (groups) and does not handle continuous variables unless they are categorized.

Cox model works for both categorical and numerical (continuous) data, allowing for more complex analysis with covariates affecting survival.

So, the Cox model is more versatile when you have a mix of categorical and numerical variables.

## 1) Plotting the Kaplan-Meier Curve

```python
kmf = KaplanMeierFitter()

kmf.fit(data.time, data.DEATH_EVENT, label = 'Kaplan Meier Estimate, full sample')

kmf.plot(linewidth=4, figsize=(12, 6))
plt.title('Heart Failure Survival: Kaplan-Meier Curve')
plt.xlabel('Tenure (Weeks)')
plt.ylabel('Survival probability')
```

```python
kmf.survival_function_  # timeline = tenure ( x axis )  # Kaplan Meier Estimate, full sample = survival probability (y axis)
```


The Kaplan-Meier model provides a table of the cumulative survival probability by time. Such "life tables" are commonly used by life insurance companies in planning life insurance products.

## Examining Varibles

A Kaplan-Meier plot can also be used to analyze impact of categorical features on survival. For example, is there a decrease in survival for patients with anaemia vs patients without anaemia?

As we continue to examine the survival function, we may want to relate survival risk with features, or characteristics of our patients. In this example, we will look at diffences in survival risk for patients who have anemia, vs. patients who don't have anemia. We can start by plotting a simple histogram for each category.

### Anemia

```python
df1 = data[data.anaemia==1]
df2 = data[data.anaemia==0]
fig, ax = plt.subplots(1,2, figsize=(14,6))
df1.groupby('DEATH_EVENT')['time'].plot(kind='hist', ax=ax[0], title='Patients with anaemia')
ax[0].legend(labels=['Censored', 'Death'])
df2.groupby('DEATH_EVENT')['time'].plot(kind='hist', ax=ax[1], title='Patients without anaemia')

## y axis : frequency of deaths  ( total deaths / total patients taken into population for case study)
# x axis: weeks (time line)
```

```python
kmf.fit(df1.time, df1.DEATH_EVENT)
kmf.plot(label='With Anaemia', figsize=(12, 6))
kmf.fit(df2.time, df2.DEATH_EVENT)
kmf.plot(label='Without Anaemia')
plt.title('Anaemia and Death: Kaplan-Meier Curve')
plt.xlabel('Weeks')
plt.ylabel('Survival probability')
```

### Blood Pressure

```python
df1 = data[data.high_blood_pressure==1]
df2 = data[data.high_blood_pressure==0]
fig, ax = plt.subplots(1,2, figsize=(16,8))
df1.groupby('DEATH_EVENT')['time'].plot(kind='hist', ax=ax[0], title='Patients with high blood pressure')
ax[0].legend(labels=['Censored', 'Death'])
df2.groupby('DEATH_EVENT')['time'].plot(kind='hist', ax=ax[1], title='Patients without high blood pressure')
```

```python
kmf.fit(df1.time, df1.DEATH_EVENT)
kmf.plot(label='With high blood pressure', figsize=(12, 6))
kmf.fit(df2.time, df2.DEATH_EVENT)
kmf.plot(label='Without high blood pressure')
plt.title('Blood Pressure and Death: Kaplan-Meier Curve')
plt.xlabel('Weeks')
plt.ylabel('Survival probability')
```

Unsurprisingly, the above plots shows us that patients with hypertension or anaemia have a increased risk of death. However, as demonstrated in the Kaplan Meier curves, hypertension and anaemia do carry increased risk of death but the confidence intervals are wide enough that these are not statistically significant.

Going back to the life insurance example, insurance companies can use such simple models to design business rules for life insurance products. For a smoker, their insurance premiums will be adjusted upwards to counteract the lowered survival probability.

## 2) Cox Proportional Hazards Model

Kaplan-Meier curves are useful only when the predictor variable is categorical (e.g.: treatment A vs treatment B; males vs females). They don’t work easily for quantitative predictors such as gene expression, weight, or age.

An alternative method is the Cox proportional hazards regression analysis, which works for both quantitative predictor variables and for categorical variables. Furthermore, the Cox regression model extends survival analysis methods to assess simultaneously the effect of several risk factors on survival time.

```python
# Using Cox Proportional Hazards Model
cph = CoxPHFitter()
cph.fit(data, duration_col='time', event_col='DEATH_EVENT')
```

### Risk factors
Lets plot the individual risk factors

```python
cph.plot()
```

## Interpretation of cph.plot():
Graph:

The plot is a visual representation of the regression coefficients (log-hazard ratios) estimated by the Cox model for each covariate.

Each covariate will be plotted on the x-axis as a point with a confidence interval shown as a horizontal line around it.

X-Axis (Log-Hazard Ratio):
The x-axis represents the log of the hazard ratio.
A positive coefficient (to the right of 0) implies an increase in the hazard (i.e., a higher risk of the event happening sooner).
A negative coefficient (to the left of 0) suggests a decrease in the hazard (i.e., a lower risk of the event).

Y-Axis (Covariates):
The y-axis lists the covariates (variables) used in the model.
Interpretation of Confidence Intervals:

If the confidence interval for a covariate crosses the vertical line at 0 (no effect), the variable may not be statistically significant.

The narrower the confidence interval, the more precise the estimate.
Example Interpretation:

A covariate with a log-hazard ratio of 1 increases the risk of the event occurring (e.g., death, failure).

A covariate with a log-hazard ratio of -1 decreases the risk of the event occurring.

Covariates with confidence intervals that do not overlap 0 can be considered significant predictors of survival.

## Obesrevation

Here, we can see the coefficient estimate, along with error bars used to assess statistical significance. We see that high blood pressure, anaemia, serum creatinine, age and ejection fraction are all within the 95% confidence interval of affecting HF death.

Simply put, older patients with high blood pressure, anaemia, serum creatine are more likely to die of heart failure. Whereas, higher ejection fraction (Percentage of blood leaving the heart at each contraction) is associated with HF survival.

### Evaluation

The most common way to measure the performance of a Cox model is to use the Concordance index or C-index. The C-index for a survival model can be thought of as the weighted average of the area under time-specific Receiver Operating Characteristic (ROC) curves.

```python
cph.print_summary(columns=["coef","exp(coef)","exp(coef) lower 95%","exp(coef) upper 95%", "z", "p"], decimals=3)
```

In the above table, the key output is exp(coef). This is interpreted as the scaling of hazard risk for each additional unit of the variable, 1.00 being neutral.

For example, the exp(coefficient) corresponding to high_blood_pressure predicts a individual with hypertension is 1.6 more times more likely to die from heart failure.
